```python
import json
```

```python
with open('./song.json', 'r') as x_path:
  song_data = json.load(x_path)
```

```python
new_song_data = []
for song in song_data:
  new_song = {}
  new_song['spotify_id'] = song['spotify_id']
  new_song['song_name'] = song['song_name']
  new_song['artist'] = song['artist']
  new_song['genius_id'] = song['genius_id']
  new_song['genre'] = song['genre']
  new_song['audio'] = {
    "valence": song['valence'],
    "arousal": song['arousal']
  }
  new_song_data.append(new_song)
```

```python
with open('songs_audio.json', 'w', encoding ='utf8') as json_file:
  json.dump(new_song_data, json_file, ensure_ascii = False)
```

```python
def reverse_audio_emotion(valence, arousal):
  # Calculate probabilities from valence and arousal
  prob_happiness = (valence + 1)/2 + (arousal + 1)/2
  prob_calm = (valence + 1)/2 - (arousal + 1)/2
  prob_anger = -(valence + 1)/2 + (arousal + 1)/2
  prob_sadness = -(valence + 1)/2 - (arousal + 1)/2

  print(prob_happiness)
  print(prob_calm)
  print(prob_anger)
  print(prob_sadness)

  # Normalize probabilities
  total_prob = prob_happiness + prob_calm + prob_anger + prob_sadness
  print(total_prob)
  prob_happiness /= total_prob
  prob_calm /= total_prob
  prob_anger /= total_prob
  prob_sadness /= total_prob

  return prob_happiness, prob_calm, prob_anger, prob_sadness

valence = 0.5
arousal = -0.2

prob_happiness, prob_calm, prob_anger, prob_sadness = reverse_audio_emotion(valence, arousal)
print(f"Probabilities of emotions:\nHappiness: {prob_happiness}\nCalm: {prob_calm}\nAnger: {prob_anger}\nSadness: {prob_sadness}")
```
